> ### Resizing & Padding

```python
# os.mkdir('./squared')

def make_square(im, min_size=256, fill_color=(0, 0, 0, 0)):
    im.convert('RGB')
    x, y = im.size
    size = max(min_size, x, y)
    new_im = Image.new(im.mode, (size, size), fill_color)
    new_im.paste(im, (int((size - x) / 2), int((size - y) / 2)))
    new_im = new_im.resize((256, 256))
    return new_im

dir_path = './temp'
# img_files = [f for f in listdir(dir_path) if isfile(join(dir_path, f))]
img_files = [f for f in listdir(dir_path)]
img_files

# 파일명을 전부 불러와서 리스트로 저장한 후 for문으로 펑션콜해서 작업 수행

for idx, img_file in enumerate(img_files):
    try:
        test_image = Image.open('./temp/' + img_file)
        new_image = make_square(test_image)
        new_image.save(f'./pet_done/{idx}.jpg')
    except Exception as e:
        print(Exception)
        print(img_file)
```
